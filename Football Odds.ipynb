```python
import os
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
df_info = pd.read_csv("dataset_final.csv")
df_info.head()
```

```python
#New DataFrame of EventID, Result (from _info) Probability of that Result from Odds

df_results1 = df_info.merge(df_odds)
```

```python
df_results = df_results1[['eventId',"result","PR_1","PR_2","PR_X"]]
df_results.head()
```

```python
list_1 = []
for row in df_results['result']:
  list_1.append('1')

df_results['list_1'] = list_1
```

```python
#Bookie predication attempt no 4

#df['dummyVar '] = df['x'].where((df['x'] > 100) & (df['y'] < 50), df['y'])

df_results['bookie_correct_loss'] = df_results['list_1'].where((df_results['PR_2']>df_results['PR_1'])&( df_results['PR_2']>df_results['PR_X'])&(df_results['result']=="2"))
df_results['bookie_correct_win'] = df_results['list_1'].where((df_results['PR_1']>df_results['PR_2'])&( df_results['PR_1']>df_results['PR_X'])&(df_results['result']=="1"))
df_results['bookie_correct_draw'] = df_results['list_1'].where((df_results['PR_X']>df_results['PR_2'])&( df_results['PR_X']>df_results['PR_1'])&(df_results['result']=="X"))
```

```python
#Bookie success dataframe attempt 1
def bookie_correct(row):
    if row['bookie_correct_loss'] =="1":
        return '1'
    if row['bookie_correct_win'] == "1":
        return '1'
    if row['bookie_correct_draw'] =="1":
        return '1'
    else:
        return '0'
        
df_results["bookie_correct"]=df_results.apply (lambda row: bookie_correct(row), axis=1)
```

```python
df_results.head()
```

## Now I have a dataframe from which I can see all the times when Bookies were accurate / not accurate in their predications

```python
df_bookie_accuracy = df_results[['eventId',"result","PR_1","PR_2","PR_X","bookie_correct"]]
df_bookie_accuracy.head()
```

```python
# First index of tuple returned by shape contains the number of index/row in dataframe
numOfRows1 = df_bookie_accuracy.shape[0]
print('Number of Rows in dataframe : ' , numOfRows1)

# Get a bool series representing which row satisfies the condition i.e. True for
# row in which value of 'Age' column is more than 30
seriesObj = df_bookie_accuracy.apply(lambda x: True if x['bookie_correct'] =="1" else False , axis=1)
# Count number of True in series
numOfRows2 = len(seriesObj[seriesObj == True].index)
print('Number of Rows in dataframe in which bookie_correct : ', numOfRows2)

print("Bookie Accuracy % :", 100*(numOfRows2 /numOfRows1))
```
